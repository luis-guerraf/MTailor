# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import numpy as np
import torchvision
```

```python
class CustomHead(nn.Module):
    def __init__(self, in_channels, num_classes):
        super().__init__()
        self.cat_or_dog = nn.Linear(in_channels, 3)     # Cat, dog or none
        self.cls_score = nn.Linear(in_channels, num_classes)
        # self.bbox_pred = nn.Linear(in_channels, num_classes * 4)
        self.bbox_pred = nn.Linear(in_channels, 4)

    def forward(self, x):
        if x.dim() == 4:
            assert list(x.shape[2:]) == [1, 1]
        x = x.flatten(start_dim=1)
        cat_or_dog = self.cat_or_dog(x)
        scores = self.cls_score(x)
        bbox_deltas = self.bbox_pred(x)

        return cat_or_dog, scores, bbox_deltas


class Model(nn.Module):

  def __init__(self):
    super(Model, self).__init__()
    ### Initialize the required Layers
    model = torchvision.models.resnet18()
    num_classes = 10    # 3 dogs + 6 cats + None
    model.fc = CustomHead(model.fc.in_features, num_classes)
    self.model = model

    self.have_object = None
    self.cat_or_dog = None
    self.specie = None
    self.bbox = None
    ### Initialize the required Layers

  def forward(self, input):
      ### Write Forward Calls for the Model
      cat_or_dog, class_scores, bbox = self.model(input)

      return {
          "cat_or_dog": cat_or_dog,
          "class_scores": class_scores,
          "bbox": bbox,
      }
```

```python
train_list = np.load('/content/assessment_dataset/train_list.npy', allow_pickle=True).tolist()
val_list = np.load('/content/assessment_dataset/val_list.npy', allow_pickle=True).tolist()
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python
from PIL import Image
import os

import torchvision.transforms as transforms


class CustomDataset(torch.utils.data.Dataset):

  def __init__(self, dataset_path, images_list, train=False):
    self.root = dataset_path
    if train:
      self.image_transforms = transforms.Compose([transforms.Resize([256, 256]),
                                                  transforms.PILToTensor(),
                                                  transforms.Lambda(lambda x: x.float()),
                                                  transforms.Normalize((127, 127, 127), (256, 256, 256))])
      self.label_transforms = lambda x: x
    else:
      self.image_transforms = transforms.Compose([transforms.Resize([256, 256]),
                                                  transforms.PILToTensor(),
                                                  transforms.Lambda(lambda x: x.float()),
                                                  transforms.Normalize((127, 127, 127), (256, 256, 256))])
      self.label_transforms = lambda x: x

    self.images_list = list(images_list)

    # self.image_label_data = read_xml_file("path_to_xml_file") # Read them all before and store in cache for faster access

  def __getitem__(self, index):
    image_name = os.path.join(self.root, f"images/{self.images_list[index]}")
    if os.path.isfile(image_name + ".jpg"):
      image = Image.open(image_name + ".jpg").convert("RGB")
    elif os.path.isfile(image_name + ".jpeg"):
      image = Image.open(image_name + ".jpeg").convert("RGB")
    else:
      image = Image.new('RGB', (256, 256))

    label = os.path.join(self.root, f"labels/{self.images_list[index]}.xml")
    if os.path.isfile(label):
      label = read_xml_file(label)
    else:
      label = {
        "cat_or_dog": "None",
        "xmin": 0,
        "ymin": 0,
        "xmax": 0,
        "ymax": 0,
        "specie": "None"
    }

    w, h = image.size
    image = self.image_transforms(image)
    label = self.label_transforms(label)
    label["xmin"] /=  w/256.0
    label["xmax"] /=  w/256.0
    label["ymin"] /=  h/256.0
    label["ymax"] /=  h/256.0

    return image, label


  def __len__(self):
      return len(self.images_list)

training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list, train=True)
training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=20, shuffle=True, num_workers=0)
images, labels = next(iter(training_loader))
# print(images)
# print(labels)

tags = {"cat":0, "dog":1, "None":2}
tags = [tags[l] for l in labels["cat_or_dog"]]
tags_species = {"basset_hound":0, "Birman":1, "pomeranian":2, "american_pit_bull_terrier":3, "american_bulldog":4,
                "Abyssinian": 5, "beagle":6, "Persian":7, "chihuahua":8, "None":9}
tags_species = [tags_species[l] for l in labels["specie"]]

# print(tags)
# print(tags_species)
```

```python
################################## HELPER CODE PROVIDED BY HIRING TEAM ##################################
def train(epochs):
  # Initialize Model and Optimizer
  model = Model().cuda()
  optimizer = torch.optim.Adam(model.parameters())

  # Initialize Loss Functions
  # have_object_loss = torch.nn.CrossEntropyLoss()    # Along with cat_or_dog
  specie_loss = torch.nn.CrossEntropyLoss()
  cat_or_dog_loss = torch.nn.CrossEntropyLoss()
  bbox_loss = torch.nn.MSELoss()

  training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list, train=True)
  training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=64, shuffle=True, num_workers=0)

  def train_one_epoch(epoch_index):
      running_loss = 0.
      last_loss = 0.

      # Here, we use enumerate(training_loader) instead of
      # iter(training_loader) so that we can track the batch
      # index and do some intra-epoch reporting
      for i, data in enumerate(training_loader):
          # Every data instance is an input + label pair
          inputs, labels = data
          inputs = inputs.cuda()

          # Make predictions for this batch
          outputs = model(inputs)

          # Compute the loss and its gradients
          tags = {"cat":0, "dog":1, "None":2}
          tags = [tags[l] for l in labels["cat_or_dog"]]
          tags_species = {"basset_hound":0, "Birman":1, "pomeranian":2, "american_pit_bull_terrier":3, "american_bulldog":4,
                          "Abyssinian": 5, "beagle":6, "Persian":7, "chihuahua":8, "None":9}
          tags_species = [tags_species[l] for l in labels["specie"]]

          loss_cat_or_dog = cat_or_dog_loss(outputs["cat_or_dog"], torch.LongTensor(tags).cuda())
          loss_specie = specie_loss(outputs["class_scores"], torch.LongTensor(tags_species).cuda())

          # gt_bbox = torch.zeros([inputs.size(0), 10, 4], device=inputs.device)
          # gt_bbox[:, tags_species, 0] = labels["xmin"].cuda().float()
          # gt_bbox[:, tags_species, 1] = labels["xmax"].cuda().float()
          # gt_bbox[:, tags_species, 2] = labels["ymin"].cuda().float()
          # gt_bbox[:, tags_species, 3] = labels["ymax"].cuda().float()
          # loss_bbox = bbox_loss(outputs["bbox"].view([inputs.size(0), 10, 4]), gt_bbox)
          gt_bbox = torch.zeros([inputs.size(0), 4], device=inputs.device)
          gt_bbox[:, 0] = labels["xmin"].cuda().float()
          gt_bbox[:, 1] = labels["xmax"].cuda().float()
          gt_bbox[:, 2] = labels["ymin"].cuda().float()
          gt_bbox[:, 3] = labels["ymax"].cuda().float()
          loss_bbox = bbox_loss(outputs["bbox"].view([inputs.size(0), 4]), gt_bbox)

          loss = loss_cat_or_dog + loss_specie + loss_bbox # Consolidate all individual losses

          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

          # Gather data and report
          running_loss += loss.item()
          if i % 10 == 0:
              last_loss = running_loss / 10 # loss per batch
              running_loss = 0.
      return last_loss

  for i in range(epochs):
    epoch_loss = train_one_epoch(i)
    print(f' Epoch {i} Loss : {epoch_loss}')

    torch.save(model.state_dict(), "model.pth")
    # metrics = test(model)
    # print(metrics)

train(100)
```
